```r
library(baQFA)
```

```r
create_and_save_dataframe <- function(num_rows, num_columns, library_name, plate_number, file_path) {
  # Generate row and column values
  
  orf <- c("edge_1", "edge_2", "edge_3", "edge_4", "edge_5", "edge_6",
           "edge_7", "sample_1", "sample_2", "sample_3", "sample_4", "edge_8",
           "edge_9", "sample_5", "sample_6", "sample_7", "sample_8", "edge_10",
           "edge_11", "sample_9", "sample_10", "sample_11", "sample_12", "edge_12",
           "edge_13", "sample_13", "sample_14", "sample_15", "sample_16", "edge_14",
           "edge_15", "sample_17", "sample_18", "sample_19", "sample_20", "edge_16",
           "edge_17", "sample_21", "sample_22", "sample_23", "sample_24", "edge_18",
           "edge_19", "edge_20", "edge_21", "edge_22", "edge_23", "edge_24"
           )
  
  

  rows <- rep(1:num_rows, times = num_columns)
  columns <- rep(1:num_columns, each = num_rows)
  
  # Create a data frame
  df <- data.frame(
    Library = rep(library_name, num_rows * num_columns),
    ORF = orf,
    Plate = rep(plate_number, num_rows * num_columns),
    Row = rows,
    Column = columns,
    Notes = character(num_rows * num_columns)
  )
  
  # Save the data frame as a TSV file
  write.table(df, file = file_path, sep = "\t", row.names = FALSE, quote = FALSE)
}

```

```r
indir_path_base <- "/mnt/g/qfa/QFA1705484730_2024-01-17_18-45-30"
outdir_path_base <- "/mnt/d/200_GitHub_Repository/OT2_yeast/results/240415_baQFA"
setwd(outdir_path_base)
num_rows <- 6
num_columns <- 8
library_name <- "Library1"
plate_number <- 1
file_path <- "LibraryDescriptions.tsv"
home_dir <- getwd()

create_and_save_dataframe(num_rows, num_columns, library_name, plate_number, file_path)
```

```r
indir_path_1 <- paste(indir_path_base, "plate1/rotate_crop/Output_Data",sep = "/")
indir_path_2 <- paste(indir_path_base, "plate2/rotate_crop/Output_Data",sep = "/")
libraries <- paste(home_dir, "LibraryDescriptions.tsv", sep = "/")

dir.create(paste(home_dir, "baQFA_plate1", sep = "/"))
dir.create(paste(home_dir, "baQFA_plate2", sep = "/"))
```

```r
qfa_table<- colonyzer.read(
  path = indir_path_1,
  libraries = libraries,
  Growth = "Intensity",
  # experiment = NA,
  # ORF2gene = NA,
  # screenID = "QFA00001"
)
```

```r
path_outdir <- paste(home_dir, "/baQFA_plate1", sep = "")
setwd(path_outdir)
# Gompertz

# GmpFit
GmpFit = qfa.fit2(qfa_table, inocguess=NULL, detectThresh=0, globalOpt=F, AUCLim=NA, TimeFormat="h", Model="Gmp")
# Construct fitness measures
GmpFit = makeFitness2(GmpFit, AUCLim=NA, plotFitness="All", filename="Gmp_fitness.pdf")
# Create plot
qfa.plot2("Gmp_GrowthCurves.pdf", GmpFit, qfa_table)
# Write CSV
write.csv(GmpFit, "Gmp_fitness.csv")

# Generalized Logistic
GlogFit = qfa.fit2(qfa_table, inocguess=NULL, detectThresh=0, globalOpt=F, AUCLim=NA, TimeFormat="h", Model="Glog")
# Construct fitness measures
GlogFit = makeFitness2(GlogFit, AUCLim=NA, plotFitness="All", filename="Glog_fitness.pdf")
# Create plot
qfa.plot2("Glog_GrowthCurves.pdf", GlogFit, qfa_table)
# Write CSV
write.csv(GlogFit, "Glog_fitness.csv")

# Standard Logistic
SlogFit = qfa.fit2(qfa_table, inocguess=NULL, detectThresh=0, globalOpt=F, AUCLim=NA, TimeFormat="h", Model="Slog")
# Construct fitness measures
SlogFit = makeFitness2(SlogFit, AUCLim=NA, plotFitness="All", filename="Slog_fitness.pdf")
# Create plot
qfa.plot2("Slog_GrowthCurves.pdf", SlogFit, qfa_table)
# Write CSV
write.csv(SlogFit, "Slog_fitness.csv")



write.csv(qfa_table, "qfa_table.csv")
```
